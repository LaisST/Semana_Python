# Problema

* Buscar dados de valores das ações;
* Criar as análises solicitadas pelo gestor;
* Enviar um email com os resultados das análises;

# Passo 1 - Buscar dados

```python
!pip install yfinance
```

```python
import yfinance
```

```python
ticker = input("Digite o código da ação: ")
dados = yfinance.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

# Passo 2 - Criar análises
- Cotação Máxima
- Cotação Mínima
- Cotação atual

```python
cotacao_maxima = fechamento.max()
cotacao_minima = fechamento.min()
cotacao_atual = fechamento[-1]

print(cotacao_maxima)
print(cotacao_minima)
print(cotacao_atual)
```

# Passo 3 - Enviar e-mail

- Abrir uma nova aba e digitar o endereço do gmail
- Clicar no botão escrever
- Preencher o destinatário
- Preencher o assunto do email 
- Escrever o email com os resultados da análise
- Clicar em enviar

```python
!pip install pyautogui
```

```python
!pip install pyperclip
```

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 3

#Abrir nova aba do gmail
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

#Clicar em escrever
pyautogui.click(x=27, y=164)

#Preencher destinatário
pyperclip.copy("laiscostast2@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")


#Preencher assunto
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#Escrever o email
mensagem = f"""Bom dia, 

Seguem as análises dos últimos 6 meses da ação {ticker}:

Cotação máxima: R$ {round(cotacao_maxima, 2)}
Cotação mínima: R$ {round(cotacao_minima, 2)}
Cotação atual: R$ {round(cotacao_atual, 2)}

Qualquer dúvida estou à disposição!

Att
Lais Costa"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

#Clicar em enter
pyautogui.click(x=802, y=694)
```
