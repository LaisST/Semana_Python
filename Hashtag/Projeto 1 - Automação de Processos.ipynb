```python
import pandas as pd
```

```python
import pyperclip
```

## Passo a passo
- Passo 1: Entrar no sistema da empresa (no link)
- Passo 2: Fazer login
- Passo 3: Exportar a base de dados
- Passo 4: Calcular os indicadores
- Passo 5: Enviar o arquivo para o chefe

```python
#Passo 1: Entrar no sistema da empresa

pyautogui.PAUSE = 1 #Pausa de 3 segundos entre os comandos

pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.hotkey("enter")
pyautogui.press("tab")

#Passo 2: Fazer login

pyperclip.copy("meu_login@gmail.com") #Login
pyautogui.hotkey("ctrl", "v") 
pyautogui.press("tab")
pyperclip.copy("1234") #Senha
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")
pyautogui.press("enter")

time.sleep(3) #Pausa de 5 segundos 

#Passo 3: Exportar base de dados
pyautogui.click(x=374, y=415) #Seleciona o arquivo
pyautogui.click(x=1721, y=185) #Clica no menu de configurações
time.sleep(2) #Pausa de 3 segundos 
pyautogui.click(x=1491, y=593) #Clica em download
time.sleep(3) #Pausa de 3 segundos 

```

```python
time.sleep(5) #Pausa de 5 segundos 
print(pyautogui.position()) #Coletar as coordenadas do mouse
```

```python
#Passo 4: Calcular os indicadores
tabela = pd.read_csv(r"C:\Users\laisc\Downloads\Compras.csv", sep=";")
display(tabela) #Mostra a tabela

total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
```

```python
#Passo 5: Enviar o arquivo para o chefe
#Entrar no email (gmail.com)

pyautogui.PAUSE = 1 #Pausa de 3 segundos entre os comandos

pyautogui.hotkey("ctrl", "t")
pyautogui.write("gmail.com")
pyautogui.hotkey("enter")

time.sleep(5)

#Clicar no botão escrever
pyautogui.click(x=21, y=195)

#Preencher as informações do email (destinatario, assunto)
pyperclip.copy("laiscostast2@gmail.com") #Escrever destinatario
pyautogui.hotkey("ctrl", "v")

pyautogui.press("tab") #proximo campo
pyperclip.copy("Relatório de vendas") #Escrever o assunto
pyautogui.hotkey("ctrl", "v")

mensagem = f'''
Prezados, 

Segue o relatório de compras

Total Gasto: R$ {total_gasto:,.2f}
Quantidade: {quantidade:,.2f}
Preço Médio: R$ {preco_medio:,.2f}

Qualquer dúvida estou a disposição
Atenciosamente 
Lais
'''

pyautogui.press("tab") #proximo campo
pyperclip.copy(mensagem) #Corpo
pyautogui.hotkey("ctrl", "v")

pyautogui.hotkey("ctrl", "enter") #Enviar email


```
