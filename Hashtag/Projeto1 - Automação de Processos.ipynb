## Passo a passo
- Passo 1: Entrar no sistema da empresa (no link)
- Passo 2: Fazer login
- Passo 3: Exportar a base de dados
- Passo 4: Calcular os indicadores
- Passo 5: Enviar o arquivo para o chefe

```python
#Passo 1: Entrar no sistema da empresa

pyautogui.PAUSE = 1 #Pausa de 3 segundos entre os comandos

pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.hotkey("enter")
pyautogui.press("tab")

#Passo 2: Fazer login

pyautogui.write("meu_login@gmail.com") #Login
pyautogui.press("tab")
pyautogui.write("1234") #Senha
pyautogui.press("tab")
pyautogui.press("enter")

time.sleep(3) #Pausa de 5 segundos 

#Passo 3: Exportar base de dados
pyautogui.click(x=374, y=415) #Seleciona o arquivo
pyautogui.click(x=1109, y=162) #Clica no menu de configurações
time.sleep(2) #Pausa de 3 segundos 
pyautogui.click(x=884, y=567) #Clica em download
time.sleep(3) #Pausa de 3 segundos 

```

```python
#Passo 4: Calcular os indicadores
tabela = pd.read_csv(r"C:\Users\cs363754\Downloads\Compras.csv", sep=";")
display(tabela) #Mostra a tabela

total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
```
