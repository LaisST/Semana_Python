```python
#imports

import pandas as pd
import plotly.express as px
```

```python
#Passo 1: Importar a base de dados
# Ajustar os caracteres para pt-br -> encoding='latin'
# Separar por ; -> sep=';'
tabela = pd.read_csv("clientes.csv", encoding='latin', sep=';')
```

```python
#Passo 2: Visualizar a base de dados
	#Entender as informações que temos disposiveis
	#Procurar bugs na base de dados

#Deletar coluna inútil
#axis=0 -> se for linha ; axis=1 -> se for coluna
tabela = tabela.drop("Unnamed: 8", axis=1)

#Exibir tabela
display(tabela)
```

```python
#Passo 3: Tratamento dos dados
#print(tabela.info())

	# Acertar informações que estão sendo reconhecidas de forma errada
    #transforma a coluna Salario Anual em número
    #Trata os erros, caso encontre algum caracter ele vai forçar transformar em um número
tabela["Salário Anual (R$)"] = pd.to_numeric(tabela["Salário Anual (R$)"], errors="coerce")
#print(tabela.info())

#Mostrar linhas vazias na coluna Profissão
#display(tabela[tabela["Profissão"].isna()])

	# Corrigir informações vazias
    #Exclui todas as linhas onde tem alguma coluna vazia
tabela = tabela.dropna()
print(tabela.info())
```

```python
#Passo 4: Análise Inicial -> Entender as notas dos clientes

display(tabela.describe()) #Detalha as informações da tabela

#Quanto maior o salário maior a nota
#clientes de promoção tem notas piores

	#Cria o gráfico
coluna ="Profissão"
grafico = px.histogram(tabela, x=coluna, y="Nota (1-100)", color="Origem", histfunc="avg", text_auto=True)

	#Exibe o gráfico
grafico.show()
```

```python
#Passo 5: Análise COmpleta -> Entender como cada característica do cliente impacta na nota
for coluna in tabela.columns:
    grafico = grafico = px.histogram(tabela, x=coluna, y="Nota (1-100)", color="Origem", histfunc="avg", text_auto=True)
    grafico.show()
```
